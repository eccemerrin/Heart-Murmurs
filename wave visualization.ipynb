```python
import os
import numpy as np
import pandas as pd
import warnings
import librosa
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
AV_path =  "/Users/ecem/Desktop/phonocardiogram/data/AV/present"
MV_path =  "/Users/ecem/Desktop/phonocardiogram/data/MV/present"
PV_path =  "/Users/ecem/Desktop/phonocardiogram/data/PV/present"
TV_path =  "/Users/ecem/Desktop/phonocardiogram/data/TV/present"
```

```python
waves_av = np.load(AV_path + "/present_AV_waves_10sec.npy", allow_pickle= True)
waves_mv = np.load(MV_path + "/present_MV_waves_10sec.npy", allow_pickle= True)
waves_pv = np.load(PV_path + "/present_PV_waves_10sec.npy", allow_pickle= True)
waves_tv = np.load(TV_path + "/present_TV_waves_10sec.npy", allow_pickle= True)
```

```python
import matplotlib as mpl
norm = mpl.colors.Normalize(vmin=0, vmax=1)
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.Blues)
cmap.set_array([])
```

```python
cmap.to_rgba()
```

```python
cmap.to_rgba(2)
```

```python
plt.rcParams.update({'font.size': 30})
fig, axs = plt.subplots(4, sharex = True, figsize = (40,30), sharey = True)
fig.suptitle("Waveforms For 10 sec Long Interval", fontsize=50 )
axs[0].plot(waves_av[0], linewidth=7, label = "AV", color = "#1C86EE")
axs[0].legend(loc="upper right", fontsize = 30) 
axs[1].plot(waves_mv[0], linewidth=7, label = "MV", color =  "#3A5FCD")
axs[1].legend(loc="upper right", fontsize = 30) 
axs[2].plot(waves_pv[0], linewidth=7, label = "PV", color = "#1874CD")
axs[2].legend(loc="upper right", fontsize = 30)
axs[3].plot(waves_tv[0], linewidth=7, label = "TV", color = "#104E8B")
axs[3].legend(loc="upper right", fontsize = 30) 
fig.supxlabel('Number of Samples')
fig.supylabel('Amplitude')
```
