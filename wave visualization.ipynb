```python
import os
import numpy as np
import pandas as pd
import warnings
import librosa
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
target_path = "/Users/ecem/Desktop/phonocardiogram/data/created_data"
```

```python
waves = np.load(target_path + "/waves.npy", allow_pickle= True)
```

```python
waves[10].max()
```

```python
waves[10].min()
```

```python
fig = plt.figure(figsize=(25, 3))
rows = 1
columns = 3 

fig.add_subplot(rows, columns, 1)
plt.plot(waves[10], color = 'purple')
plt.title("wave")

plt.show()


fig = plt.figure(figsize=(25, 3))
rows = 1
columns = 3 

fig.add_subplot(rows, columns, 1)
plt.plot(waves[10][100000:(22050*10+100000)], color = 'purple')
plt.title(" clipped wave ")

plt.show()
```
