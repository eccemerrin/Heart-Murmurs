```python
import os
import numpy as np
import pandas as pd

import warnings
import librosa
warnings.filterwarnings('ignore')
```

```python
dir_path = "/Users/ecem/Desktop/phonocardiogram/data/training_data"
```

```python
target_path = "/Users/ecem/Desktop/phonocardiogram/data/MV"
```

## get all files with _AV 

```python
audio_files = []
```

```python
for file in os.listdir(dir_path):
    if file.endswith("MV.wav"):
        audio_files.append(os.path.join(dir_path, file))
```

```python
audio_files = sorted(audio_files)
audio_files
```

```python
len(audio_files)
```

### if there is multiple recordings to get them

```python
audio_files_mr = [] #mr stands for multiple recording.
for file_ in os.listdir(dir_path):
    if file_.endswith("MV_2.wav"):
        audio_files_mr.append(os.path.join(dir_path, file_))
        
    
```

```python
audio_files_mr = sorted(audio_files_mr)
audio_files_mr
```

```python
len(audio_files_mr)
```

### sum all audio files

```python
audio = audio_files + audio_files_mr 
print(len(audio))
```

```python
audio
```

## now we need to split them according to their murmur status

```python
present_id = np.load(target_path + "/murmur_present_ids.npy", allow_pickle = True)
```

```python
present_id.shape
```

```python
absent_id = np.load(target_path + "/murmur_absent_ids.npy", allow_pickle = True)
```

```python
absent_id.shape
```

### check whether there is a recording with multiple recordings

```python
for i in range(len(audio_files_mr)):
    if int(audio_files_mr[i].split("/")[-1].split("_")[0]) in present_id:
        print(audio_files_mr[i].split("/")[-1].split("_")[0])
```

```python
for i in range(len(audio_files_mr)):
    if int(audio_files_mr[i].split("/")[-1].split("_")[0]) in absent_id:
        print(audio_files_mr[i].split("/")[-1].split("_")[0])
```

## now we are getting full path of murmur present av waves so that we can fed them into librosa

### for murmur present waves

```python
present_waves = []
for i in range(len(audio)):
    if (int(audio[i].split("/")[-1].split("_")[0]) in present_id):
        present_waves.append('/Users/ecem/Desktop/phonocardiogram/data/training_data/' 
                       + audio[i].split("/")[-1].split("_")[0] + '_MV.wav')
```

```python
len(present_waves)
```

```python
present_waves = sorted(present_waves)
```

```python
waves_present = []
for i in range(len(present_waves)):
    waves_present.append(np.array(librosa.load(present_waves[i])[0])) #waves
```

```python
len(waves_present)
```

### for murmur absent waves

```python
absent_waves = []
for i in range(len(audio)):
    if (int(audio[i].split("/")[-1].split("_")[0]) in absent_id) and (audio[i].split("/")[-1].split("_")[0] != str(50321)):
        absent_waves.append('/Users/ecem/Desktop/phonocardiogram/data/training_data/'
                            + audio[i].split("/")[-1].split("_")[0] + '_MV.wav')
        
absent_waves.append('/Users/ecem/Desktop/phonocardiogram/data/training_data/50321_MV_2.wav')
```

```python
absent_waves = sorted(absent_waves)
```

```python
len(absent_waves)
```

```python
waves_absent = []
for i in range(len(absent_waves)):
    waves_absent.append(np.array(librosa.load(absent_waves[i])[0])) #waves
```

```python
#sanity check
len(waves_absent)
```

### save

```python
np.save(target_path + "/present/present_MV_waves", waves_present, allow_pickle= True)
```

```python
np.save(target_path + "/absent/absent_MV_waves", waves_absent, allow_pickle= True)
```
