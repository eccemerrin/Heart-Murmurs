## manipulate the csv


### split recording locations 

```python
recording_locations = csv["Recording locations:"].str.split("+", n = 3, expand = True)
recording_locations
```

```python
recordings = np.zeros(shape=(942, 4))
```

```python
for i in range(942):
    if recording_locations.iloc[i][0] == 'AV':
        recordings[i][0] = 1
        
    if recording_locations.iloc[i][1] == 'PV':
        recordings[i][2] = 1
        
    if recording_locations.iloc[i][2] == 'TV':
        recordings[i][3] = 1
        
    if recording_locations.iloc[i][3] == 'MV':
        recordings[i][1] = 1
```

```python
recordings
```

### split murmur present locations

```python
murmur_location = csv["Murmur locations"].str.split("+",  n = 3, expand = True)
murmur_location
```

```python
murmur_0 = np.zeros(shape=(942, 4))
murmur_1 = np.zeros(shape=(942, 4))
murmur_2 = np.zeros(shape=(942, 4))
murmur_3 = np.zeros(shape=(942, 4))
```

```python
for i in range(len(csv)):
    if murmur_location.iloc[i][0] == 'AV':
        murmur_0[i][0] = 1
    if murmur_location.iloc[i][0] == 'MV':
        murmur_0[i][1] = 1
    if murmur_location.iloc[i][0] == 'PV':
        murmur_0[i][2] = 1
    if murmur_location.iloc[i][0] == 'TV':
        murmur_0[i][3] = 1
```

```python
murmur_0
```

```python
for i in range(len(csv)):
    if murmur_location.iloc[i][1] == 'AV':
        murmur_1[i][0] = 1
        
    if murmur_location.iloc[i][1] == 'MV':
        murmur_1[i][1] = 1
        
    if murmur_location.iloc[i][1] == 'PV':
        murmur_1[i][2] = 1
        
    if murmur_location.iloc[i][1] == 'TV':
        murmur_1[i][3] = 1
        
murmur_1
```

```python
for i in range(len(csv)):
    if murmur_location.iloc[i][2] == 'AV':
        murmur_2[i][0] = 1
        
    if murmur_location.iloc[i][2] == 'MV':
        murmur_2[i][1] = 1
        
    if murmur_location.iloc[i][2] == 'PV':
        murmur_2[i][2] = 1
        
    if murmur_location.iloc[i][2] == 'TV':
        murmur_2[i][3] = 1
        
murmur_2
```

```python
for i in range(len(csv)):
    if murmur_location.iloc[i][3] == 'AV':
        murmur_3[i][0] = 1
        
    if murmur_location.iloc[i][3] == 'MV':
        murmur_3[i][1] = 1
        
    if murmur_location.iloc[i][3] == 'PV':
        murmur_3[i][2] = 1
        
    if murmur_location.iloc[i][3] == 'TV':
        murmur_3[i][3] = 1
        
murmur_3
```

```python
murmur = murmur_0 + murmur_1 + murmur_2 + murmur_3
```

```python
murmur
```

### create their dataframe

```python
recording_df = pd.DataFrame(recordings, columns =['recording location AV', 'recording location MV'
                                                        , 'recording location PV', 'recording location TV'])
recording_df
```

```python
murmur_df = pd.DataFrame(murmur, columns =['murmur location AV', 'murmur location MV'
                                                        , 'murmur location PV', 'murmur location TV'])
murmur_df
```

### drop recording  locations and murmr locaitons and append new columns

```python
csv.drop(['Recording locations:', 'Murmur locations'], axis =1,  inplace = True)
```

```python
csv
```

```python
csv = pd.concat([csv, recording_df, murmur_df], axis = 1 )
csv
```

### now get index values of  the rows with Murmur == present, absent,unknown

```python
idx_unknown  = np.where(csv['Murmur'] == 'Unknown')
idx_unknown
```

```python
idx_present = np.where((csv['Murmur'] == 'Present'))
idx_present
```

```python
idx_absent = np.where((csv['Murmur'] == 'Absent'))
idx_absent
```

### get indexes for recordings taken in specific location with murmur = absent, present

```python
#this shows indexes of recordings taken in av location which also has a murmur at av
idx_av_present =  np.where((csv['Murmur'] == 'Present') & (csv['murmur location AV'] == 1) &
                           (csv['recording location AV'] == 1))
idx_av_present
```

```python
#this shows indexes of recordings taken in av location which also has not a murmur at anywhere
idx_av_absent =  np.where((csv['Murmur'] == 'Absent') & (csv['murmur location AV'] == 0) &
                           (csv['recording location AV'] == 1))
idx_av_absent
```
