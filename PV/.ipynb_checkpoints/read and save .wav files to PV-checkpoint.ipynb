```python
import os
import numpy as np
import pandas as pd
import warnings
import librosa
warnings.filterwarnings('ignore')
```

```python
dir_path = "/Users/ecem/Desktop/phonocardiogram/data/training_data"
```

```python
target_path = "/Users/ecem/Desktop/phonocardiogram/data/AV"
```

```python
audio_files = []
```

```python
for file in os.listdir(dir_path):
    if file.endswith("AV.wav"):
        audio_files.append(os.path.join(dir_path, file))
```

```python
audio_files = sorted(audio_files)
audio_files
```

```python
len(audio_files)
```

```python
audio_files_mr = [] #mr stands for multiple recording.
for file_ in os.listdir(dir_path):
    if file_.endswith("AV_3.wav"):
        audio_files_mr.append(os.path.join(dir_path, file_))
        
    elif file_.endswith("AV_2.wav") and not file_.startswith("50743_AV"):
        audio_files_mr.append(os.path.join(dir_path, file_))
        
    
```

```python
audio_files_mr = sorted(audio_files_mr)
audio_files_mr
```

```python
len(audio_files_mr)
```

```python
audio = audio_files + audio_files_mr 
print(len(audio))
```

```python
waves_array = []
```

```python
for i in range(len(audio)):
    waves_array.append(np.array(librosa.load(audio[i])[0])) #waves
```

```python
len(waves_array)
```

```python
np.save(target_path + "/AV_waves", waves_array, allow_pickle= True)
```
